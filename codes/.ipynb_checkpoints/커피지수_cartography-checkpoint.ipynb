## <a name="comparison-calculation"></a>  II. 커피 브랜드별 비교 지수
    >> 비교 지수 계산시, 각 브랜드의 매장수는 해당 브랜드의 전국 매장수에 대해 각 지역이 얼마의 비중을 차지하는가(비례값, 이하 기본 지수라고 명한다)를 계산한 지수로 대체하였다. 이는 브랜드별 전국 매장수가 최대 2300개 이상(이디야)에서 최소 153개(커피빈)까지 차이가 나는 점과 매장 수가 본사의 규모 및 투자크기 등 고려하기 어려운 변수에 크게 영향을 받을 것을 감안한 것이다.
기본 지수 계산 
지수_pk
지역별 빽다방 매장 수 / 전국 빽다방 매장 수
          </li>            <li style = "font-family: Gulim;font-size:16px;line-height:1.5">
              <b>지수_ed</b> <br>
              지역별 이디야 매장 수 / 전국 이디야 매장 수
          </li>              <li style = "font-family: Gulim;font-size:16px;line-height:1.5">
              <b>지수_sb</b> <br>
              지역별 스타벅스 매장 수 / 전국 스타벅스 매장 수
          </li>          <li style = "font-family: Gulim;font-size:16px;line-height:1.5">
              <b>지수_cb</b> <br>
              지역별 커피빈 매장 수 / 전국 커피빈 매장 수
          </li> 
      </ul>


```python
tot_paik = cof_m1['빽다방'].sum() ; cof_m1['지수_pk'] = cof_m1['빽다방'] /tot_paik
tot_ed = cof_m1['이디야'].sum() ; cof_m1['지수_ed'] = cof_m1['이디야'] /tot_ed
tot_sb = cof_m1['스벅'].sum() ; cof_m1['지수_sb'] = cof_m1['스벅'] /tot_sb
tot_cb = cof_m1['커피빈'].sum() ; cof_m1['지수_cb'] = cof_m1['커피빈'] /tot_cb

cof_m1.head()
```

<ul style="list-style-type:disc;">
   <li style = "font-family: Gulim; font-size:20px;line-height:2.0">
      <b> 국산 브랜드 vs. 해외 브랜드 비교 </b>
      <ul  style="list-style-type:circle;">
          <li style = "font-family: Gulim;font-size:14px;line-height:1.5">
              <b>지수_국내해외</b> <br>
              (지수_pk + 지수_ed) / (지수_sb + 지수_cb) 
          </li> 

```python
cof_m1['지수_해외국내'] =  (cof_m1['지수_sb'] + cof_m1['지수_cb'])/(cof_m1['지수_pk'] + cof_m1['지수_ed'])
cof_m1.head()
```

<ul style="list-style-type:disc;">
   <li style = "font-family: Gulim; font-size:20px;line-height:2.0">
      <b> 가격대별 브랜드 분포도 비교 </b>
      <ul  style="list-style-type:circle;">
          <li style = "font-family: Gulim;font-size:14px;line-height:1.5">
              <b>지수_최저가</b> <br>
               지수_pk / (지수_ed + 지수_sb + 지수_cb) 
          </li>  <li style = "font-family: Gulim;font-size:14px;line-height:1.5">
              <b>지수_최고가</b> <br>
               지수_cb / (지수_pk + 지수_ed + 지수_sb) 
          </li>  <li style = "font-family: Gulim;font-size:14px;line-height:1.5">
              <b>지수_고가</b> <br>
               지수_sb / (지수_pk + 지수_ed + 지수_cb) 
          </li> 

```python
cof_m1['지수_최저가'] = cof_m1['지수_pk'] / (cof_m1['지수_ed'] + cof_m1['지수_sb'] + cof_m1['지수_cb'])
cof_m1['지수_최고가'] = cof_m1['지수_cb'] / (cof_m1['지수_pk'] + cof_m1['지수_ed'] + cof_m1['지수_sb'])
cof_m1['지수_고가'] = cof_m1['지수_sb'] / (cof_m1['지수_pk'] + cof_m1['지수_ed'] + cof_m1['지수_cb'])
cof_m1.head()
```

```python
cof_m1.to_csv("data_soomin/커피지수및인구비율_euckr인코딩.csv")
cof_m1.to_csv("data_soomin/커피지수및인구비율.csv", encoding = 'utf-8-sig') 
```

```python
plt.figure(figsize=(8, 11))
# 지역 이름 표시
for idx, row in draw_korea.iterrows():
    
    # 광역시는 구 이름이 겹치는 경우가 많아서 시단위 이름도 같이 표시한다. 
    # (중구, 서구)
    if len(row['ID'].split())==2:
        dispname = '{}\n{}'.format(row['ID'].split()[0], row['ID'].split()[1])
    elif row['ID'][:2]=='고성':
        dispname = '고성'
    else:
        dispname = row['ID']

    # 서대문구, 서귀포시 같이 이름이 3자 이상인 경우에 작은 글자로 표시한다.
    if len(dispname.splitlines()[-1]) >= 3:
        fontsize, linespacing = 9.5, 1.5
    else:
        fontsize, linespacing = 11, 1.2

    plt.annotate(dispname, (row['x']+0.5, row['y']+0.5), weight='bold',
                 fontsize=fontsize, ha='center', va='center', 
                 linespacing=linespacing)
    
# 시도 경계 그린다.
for path in BORDER_LINES:
    ys, xs = zip(*path)
    plt.plot(xs, ys, c='black', lw=1.5)

plt.gca().invert_yaxis()
#plt.gca().set_aspect(1)

plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
# 인천 미추(미추홀구)의 경우 사용하고자 하는 cartogram 의 ID 에 존재하지 않으므로 삭제한다.

cof_m2 = cof_m1.drop(cof_m1[cof_m1['ID'] == '인천 미추'].index)
set(cof_m2['ID'].unique()) - set(draw_korea['ID'].unique()) 
```

```python
# cartogram을 그리기 위해 draw_korea 데이터프레임과 카페브랜드 데이터프레임을 합치도록 한다.

cof_m3 = pd.merge(cof_m2, draw_korea, how = 'left', on = ['ID'])
cof_m3.head()
```

```python
def drawKorea(targetData, blockedMap, cmapname):
    gamma = 0.75

    whitelabelmin = (max(blockedMap[targetData]) - 
                                     min(blockedMap[targetData]))*0.25 + \
                                                                min(blockedMap[targetData])

    datalabel = targetData

    vmin = min(blockedMap[targetData])
    vmax = max(blockedMap[targetData])

    mapdata = blockedMap.pivot_table(index='y', columns='x', values=targetData)
    masked_mapdata = np.ma.masked_where(np.isnan(mapdata), mapdata)
    
    plt.figure(figsize=(9, 11))
    plt.pcolor(masked_mapdata, vmin=vmin, vmax=vmax, cmap=cmapname, 
               edgecolor='#aaaaaa', linewidth=0.5)

    # 지역 이름 표시
    for idx, row in blockedMap.iterrows():
        # 광역시는 구 이름이 겹치는 경우가 많아서 시단위 이름도 같이 표시한다. 
        #(중구, 서구)
        if len(row['ID'].split())==2:
            dispname = '{}\n{}'.format(row['ID'].split()[0], row['ID'].split()[1])
        elif row['ID'][:2]=='고성':
            dispname = '고성'
        else:
            dispname = row['ID']

        # 서대문구, 서귀포시 같이 이름이 3자 이상인 경우에 작은 글자로 표시한다.
        if len(dispname.splitlines()[-1]) >= 3:
            fontsize, linespacing = 10.0, 1.1
        else:
            fontsize, linespacing = 11, 1.

        annocolor = 'white' if row[targetData] > whitelabelmin else 'black'
        plt.annotate(dispname, (row['x']+0.5, row['y']+0.5), weight='bold',
                     fontsize=fontsize, ha='center', va='center', color=annocolor,
                     linespacing=linespacing)

    # 시도 경계 그린다.
    for path in BORDER_LINES:
        ys, xs = zip(*path)
        plt.plot(xs, ys, c='black', lw=2)

    plt.gca().invert_yaxis()

    plt.axis('off')

    cb = plt.colorbar(shrink=.1, aspect=10)
    cb.set_label(datalabel)

    plt.tight_layout()
    plt.show()
```

              또한 이들은 전국에 걸친 기타 대도시(대전, 대구, 창원 등)에도 대체로 높은 밀도를 보여 전국에 고르게 퍼진 추이이다. <br><br>
                      반면 커피빈과 스타벅스의 경우 공통적으로 서울 강남을 중심으로 서울 내 매우 밀집해 있으며 근교 경기지역에서는 상대적으로 덜한 밀도를 보이고 있다. 이들은 제 2도시로 여겨지는 부산에 약간의 밀도를 보이는 것을 제외하면 서울지역을 중심으로 매우 한정된 집중도를 가지는 것으로 보인다.
              </b>
          </li> </ul>

          <li style = "font-family: Gulim;font-size:16px;line-height:1.6">
              <b>해외 vs. 국내 브랜드 </b> <br> >> 일부 지역의 inf로 인해 그래프가 올바르게 그려지지 않을 수 있어 np.inf를 np.nan 으로 변경한다. nan(그래프 상 흰 색 셀)의 경우, 해당 지역에 외국브랜드(스타벅스, 커피빈)와 비교대상인 국내 브랜드가 존재하지 않았음 으로 이해할 수 있다.<br><br>
              <b>그래프에 따르면 서울 중심지역(강남, 마포, 서초 등) 및 일부 광역시의 중심지(대구 수성구, 부산 해운대, 인천 중구, 광주 남구 등)에는 해외브랜드의 집중도가 국내브랜드에 비해 월등히 높은 것으로 보인다. 반면 대부분의 일반 지역에서는 국내브랜드의 집중도가 높다.<br>계산에는 매장 수가 아닌 전국 매장수에 대비 지역별 비례값을 사용하였으므로, 이 결과는 '해외 브랜드 매장이 다수 위치한' 지역이 아닌, 해당 해외 브랜드에서 매장 개점 및 운영에 대해 타지역에 비해 중요시하는 지역을 파악하는 것으로 한다.  
              </b>
    </li></ul><br>

```python
cof_m3_1 = cof_m3.replace([np.inf, -np.inf], np.nan)
cof_m3_1.head()
```

```python
drawKorea('지수_해외국내', cof_m3_1, 'RdBu')
```

              <b>최저가 브랜드로 뽑은 빽다방의 경우, 서울, 부산 및 대부분 대도시의 중심지역 등 발달한 지역일수록 타브랜드에 비해 밀도가 떨어지는 것으로 보인다. 반면 최고가 브랜드로 꼽힌 커피빈의 경우 정반대의 양상을 보이는데, 비교대상인 세 브랜드에 비해 서울과 부산 등의 대도시에 유의하게 밀집함을 볼 수 있다. 고가 브랜드로 조사된 스타벅스는 커피빈과 비슷하게 타지역에 비해서 서울, 부산 등 대도시에 집중하는 추이가 있으나 뚜렷하지 않고, 오히려 의외일 수 있는 고성, 여주, 과천 등에서 강세로 나타나며 대구, 광주 등 기타 광역시에 보다 집중하고 있다.
              </b>
          </li>
